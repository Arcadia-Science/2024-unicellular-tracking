# Principal components analysis of motility features
---

#### Overview

In this notebook we perform principal components analysis (PCA) to determine which features of motility are driving variation between groups of interest.

Output from this notebook was used to generate Figure 7 in the [pub](https://doi.org/10.57844/arcadia-2d61-fb05).

```python
import string
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from swimtracker.stats_testing import joint_grid_with_stats
```

Configure matplotlib styles to match the Arcadia style guide

```python
import arcadia_pycolor as apc

apc.mpl.setup()
```

## Load motility metrics
---

Load summary motility metrics from all _C. Reinhardtii_ cells swimming in pools and wells. This data is stored in `../data/AMID-04_summary_motility_metrics.csv` and `../data/AMID-05_summary_motility_metrics.csv`, respectively. Both csv files were generated by running [`1_compute-summary-motility-metrics.ipynb`](./1_compute-summary-motility-metrics.ipynb).

#### Experimental parameters

| strain  | vessel type | position_in_tube | time_in_water |
|---------|-------------|------------------|---------------|
| CC-124  | pools       | top              | 4 hrs         |
|         | wells       | middle           | 21 hrs        |

Strain [CC-124](https://www.chlamycollection.org/product/cc-124-wild-type-mt-137c/) corresponds to _C. Reinhardtii_.

#### Collect csv files

```python
csv_files = {
    "AMID-04": Path("../data/AMID-04_summary_motility_metrics.csv"),
    "AMID-05": Path("../data/AMID-05_summary_motility_metrics.csv"),
}

motility_metrics_dataframe = pd.concat(
    [pd.read_csv(csv_file) for csv_file in csv_files.values()],
)

# preview dataframe
motility_metrics_dataframe.drop("slide_ID", axis=1).groupby("experiment_ID").head(4)
```

## PCA
---

#### Define experimental variables and motility metrics

```python
# set metrics
metrics_subset = {
    "max_sprint_length": "Max sprint length",
    "confinement_ratio": "Confinement ratio",
    "mean_curvilinear_speed": "Mean curvilinear speed",
    "mean_linear_speed": "Mean linear speed",
    "mean_angular_speed": "Mean angular speed",
    "pivot_rate": "Pivot rate",
}

# set experimental variables
experimental_variables = [
    "position_in_tube",
    "time_in_water",
    "vessel_type",
]

# set color palettes in accordance with experimental variables
palettes = {
    "time_in_water": ["#73B5E3", "#377BB7"],
    "position_in_tube": ["#B5BEA4", "#596F74"],
    "vessel_type": ["#F28360", "#7A77AB"],
}
```

#### Choose number of components for PCA

Choose the number of components that captures 95% of the variance.

```python
# min-max normalization
minmax_range = (
    motility_metrics_dataframe[metrics_subset.keys()].max()
    - motility_metrics_dataframe[metrics_subset.keys()].min()
)
motility_metrics_normalized = (
    motility_metrics_dataframe[metrics_subset.keys()]
    - motility_metrics_dataframe[metrics_subset.keys()].min()
) / minmax_range

# cumulative explained variance
pca_cumulative_variance = PCA().fit(motility_metrics_normalized)
cumulative_variance = np.cumsum(pca_cumulative_variance.explained_variance_ratio_)
num_components = np.arange(cumulative_variance.size) + 1

# create figure
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(num_components, 100 * cumulative_variance, "o-", mfc="none")
ax.axhline(95, color="k", ls="--", label="95% ref line")

# aesthetics
sns.despine(ax=ax)
ax.set_xlabel("Number of components")
ax.set_ylabel("Cumulative explained variance")
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.legend(loc=4)
```

#### Build PCA

```python
# PCA
num_pca_components = num_components[cumulative_variance > 0.95][0]
pca = PCA(n_components=num_pca_components)
motility_components = pca.fit_transform(motility_metrics_normalized)

# add PCA component to dataframe
for i in range(num_pca_components):
    key = f"PC-{i + 1}"
    motility_metrics_dataframe[key] = motility_components[:, i]
```

#### PCA Weights

```python
cmap = apc.gradients.red_blue.reverse().to_mpl_cmap()

# create figure
figsize = apc.mpl.get_figure_dimensions("float_wide")
fig, ax = plt.subplots(figsize=figsize, layout="constrained")

# create heatmap
ax = sns.heatmap(
    pca.components_.T,
    cmap=cmap,
    xticklabels=[f"PC-{i}" for i in range(1, pca.n_components_ + 1)],
    yticklabels=metrics_subset.values(),
    annot=True,
    square=True,
    cbar=True,
    center=0,
    ax=ax,
)

# figure aesthetics
apc.mpl.style_plot(colorbar_exists=True)

# save figure
apc.mpl.save_figure("../results/figure-7A.svg")
```

#### PCA for each experimental variable

```python
source = motility_metrics_dataframe.copy()

# loop through experimental variables
for k, experimental_variable in enumerate(experimental_variables):
    # render joint distributions with stats
    ax = joint_grid_with_stats(
        data=source,
        x_variable="PC-1",
        y_variable="PC-2",
        hue_variable=experimental_variable,
        palette=palettes[experimental_variable],
        height=5,
    )

    # aesthetics
    variances = pca_cumulative_variance.explained_variance_ratio_
    ax.ax_joint.set_xlabel(f"PC-1  ({variances[0]:.1%})")
    ax.ax_joint.set_ylabel(f"PC-2  ({variances[1]:.1%})")
    sns.move_legend(ax.ax_joint, "upper right", bbox_to_anchor=(1.7, 1))
    apc.mpl.style_plot(ax.ax_joint, monospaced_axes="both")

    # save figure
    panel = string.ascii_uppercase[k + 1]
    apc.mpl.save_figure(f"../results/figure-7{panel}.svg")
```
